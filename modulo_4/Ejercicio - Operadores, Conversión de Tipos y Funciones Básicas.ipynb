```python
def evaluar_digitos():
    numero_uno = input('Digite el primer numero: ')
    numero_dos = input('Digite el segundo numero: ')
    if(numero_uno.isdigit() == True and numero_dos.isdigit() == True):  
            return eval(numero_uno),eval(numero_dos)
    else:
        if((numero_uno.startswith('0b') or numero_uno.startswith('0B')) and (numero_dos.startswith('0b') or numero_dos.startswith('0B'))):
            numero_uno = binario_a_decimal(numero_uno[2:])
            numero_dos = binario_a_decimal(numero_dos[2:])
        elif((numero_uno.startswith('0o') or numero_uno.startswith('0O')) and (numero_dos.startswith('0o') or numero_dos.startswith('0o'))):
            numero_uno = octal_a_decimal(numero_uno[2:])
            numero_dos = octal_a_decimal(numero_dos[2:])
        elif((numero_uno.startswith('0x') or numero_uno.startswith('0X')) and (numero_dos.startswith('0X') or numero_dos.startswith('0x'))):
            numero_uno = hexadecimal_a_decimal(numero_uno[2:])
            numero_dos = hexadecimal_a_decimal(numero_dos[2:])
        return numero_uno, numero_dos

####################################### --- Función para convertir de octal a decimal --- #######################################

def octal_a_decimal(octal):
    decimal = 0
    posicion = 0
    # Invertir octal, porque debemos recorrerlo de derecha a izquierda
    # pero for in empieza de izquierda a derecha
    octal = octal[::-1]
    for digito in octal:
        valor_entero = int(digito)
        numero_elevado = int(8 ** posicion)
        equivalencia = int(numero_elevado * valor_entero)
        decimal += equivalencia
        posicion += 1
    return decimal

####################################### --- Función para convertir de hexadecimal a decimal --- #######################################

def obtener_valor_real(caracter_hexadecimal):
    equivalencias = {
        "f": 15,
        "e": 14,
        "d": 13,
        "c": 12,
        "b": 11,
        "a": 10,
    }
    if caracter_hexadecimal in equivalencias:
        return equivalencias[caracter_hexadecimal]
    else:
        return int(caracter_hexadecimal)


def hexadecimal_a_decimal(hexadecimal):
    # Convertir a minúsculas para hacer las cosas más simples
    hexadecimal = hexadecimal.lower()
    # La debemos recorrer del final al principio, así que la invertimos
    hexadecimal = hexadecimal[::-1]
    decimal = 0
    posicion = 0
    for digito in hexadecimal:
        # Necesitamos que nos dé un 10 para la A, un 9 para el 9, un 11 para la B, etcétera
        valor = obtener_valor_real(digito)
        elevado = 16 ** posicion
        equivalencia = elevado * valor
        decimal += equivalencia
        posicion += 1
    return decimal

####################################### --- Función para convertir de binario a decimal --- #######################################

def binario_a_decimal(binario):
    posicion = 0
    decimal = 0
    # Invertir la cadena porque debemos recorrerla de derecha a izquierda
    binario = binario[::-1]
    for digito in binario:
        # Elevar 2 a la posición actual
        multiplicador = 2**posicion
        decimal += int(digito) * multiplicador
        posicion += 1
    return decimal
#########################################################################################################################################
#########################################################################################################################################
#########################################################################################################################################
#########################################################################################################################################

def creating_array():
    while True:
        try:
            lista_numeros = eval(input('Digite una lista de 5 numeros en el formato que usted guste: '))
            tipado = type(lista_numeros)
            # esto lo hacemos ya que si lo que digito el usuario es un set, tupla o lista, cualquiera de ellos se puede convertir facilmente
            # a una lista con este metodo, y si no lo digito bien, significará que la estructura no es correcta o que lo que sea que este 
            # digitando no es iterable para ser transformado a cualquiera de los 3
            lista_numeros = list(lista_numeros)
            break
        except:
            print('Lo que usted esta digitando no viene incluido en el formato de lista, set o tupla, intentelo de nuevo porfavor!')
    contador = -1
    for i in lista_numeros:
        contador += 1
        if(str(i).isdigit() == True):
            pass
        else:
            if(i.startswith('0b') or i.startswith('0B')):
                lista_numeros[contador] = binario_a_decimal(i[2:])
            elif(i.startswith('0o') or i.startswith('0O')):
                lista_numeros[contador] = octal_a_decimal(i[2:])
            elif(i.startswith('0x') or i.startswith('0X')):
                lista_numeros[contador] = hexadecimal_a_decimal(i[2:])
    return tipado(lista_numeros)
    
def operar_numeros(lista_numeros,operacion):  
    contador = 0
    if(operacion == '1'):
        for contador_dos in lista_numeros:
            contador += contador_dos
        print(f'La sumatoria de los numeros {lista_numeros} es: {contador}')
    elif(operacion == '2'):
        for contador_dos in lista_numeros:
            contador -= contador_dos
        print(f'La resta de todos los numeros de la lista {lista_numeros} es de: {contador}')
    elif(operacion == '3'):
        lista_numeros_dos = list(lista_numeros)
        contador = lista_numeros_dos[0]
        lista_numeros_dos.pop(0)
        for contador_dos in lista_numeros_dos:
            contador *= contador_dos
        print(f'La multiplicación entre todos los numeros de la lista {lista_numeros} da como respuesta: {contador}')
    elif(operacion == '4'):
        lista_numeros_dos = list(lista_numeros)
        contador = lista_numeros_dos[0]
        lista_numeros_dos.pop(0)
        for contador_dos in lista_numeros_dos:
            contador /= contador_dos
        print(f'La division desde el 1er numero de la lista {lista_numeros} hasta el numero {lista_numeros[-1]} es: {contador}')
    elif(operacion == '5'):
        lista_numeros_dos = list(lista_numeros)
        contador = lista_numeros_dos[0]
        lista_numeros_dos.pop(0)
        for contador_dos in lista_numeros_dos:
            contador **= contador_dos
        print(f'''Si vamos incrementando la potencia del 1er numero ({lista_numeros[0]}) de la lista {lista_numeros} 
hasta llegar hasta el último ({lista_numeros[-1]}),tendriamos como resultado lo siguiente: {contador}''')

answer = 'y'
while(answer == 'y' or answer == 'Y'):
    answer = input('\nDigite a) para sumar solametne dos numeros y b) para trabajar con listas de hasta 5 numeros (a/b): ').lower()
    if(answer == 'a'):
        print('''\nUtiliza 1) para sumar dos numeros, 2) para restar dos numeros, 3) para multiplicar dos numeros, 4) para dividir dos numeros y 
5) para digitar primero un numero y elevarlo a la potencia del segundo que digite''')
        answer = input('\nDigite uno de los incisos anteriormente indicados porfavor (1/2/3/4/5): ')
        if(answer == '1'):
            numero_uno = ''; numero_dos = ''
            numero_uno,numero_dos = evaluar_digitos()
            print(f'\nLa suma del numero {numero_uno} y el numero {numero_dos} es: {numero_uno+numero_dos}')
        elif(answer == '2'):
            numero_uno = ''; numero_dos = ''
            numero_uno,numero_dos = evaluar_digitos()
            print(f'\nLa resta del numero {numero_uno} menos el numero {numero_dos} es: {numero_uno-numero_dos}')
        elif(answer == '3'):
            numero_uno = ''; numero_dos = ''
            numero_uno,numero_dos = evaluar_digitos()
            print(f'\nLa multiplicación entre el numero {numero_uno} y el numero {numero_dos} es: {numero_uno*numero_dos}')
        elif(answer == '4'):
            numero_uno = ''; numero_dos = ''
            numero_uno,numero_dos = evaluar_digitos()
            print(f'\nLa division entre el numero {numero_uno} y el numero {numero_dos} es: {numero_uno/numero_dos}')
        elif(answer == '5'):
            numero_uno = ''; numero_dos = ''
            numero_uno,numero_dos = evaluar_digitos()
            print(f'\nLa potencia del numero {numero_uno} elevado a la {numero_dos} es: {numero_uno**numero_dos}')
    elif(answer == 'b'):
        lista_numeros = creating_array()
        if(len(lista_numeros) == 5):
            print('''\nUtiliza 1) para sumar una lista de 5 numeros, 2) para ir restando una lista de 5 numeros, 3) para multiplicar entre si una lista 
        de 5 numeros, 4) para dividir una lista de 5 numeros y 5) para digitar una lista de 5 numeros y elevarlos a la potencia en serie en el orden en
        que hallan digitado''')
            answer = input('Digite uno de los incisos anteriormente indicados porfavor (1/2/3/4/5): ')
            operar_numeros(lista_numeros,answer)
        else:
            print(f'Lo sentimos, la lista {lista_numeros} resultó no tener los 5 digitos esperados! ')
            pass       
    else:
        print(f'Lo sentimos, su respuesta {answer} es ¡INVÁLIDA!')
    answer = input('Desea intentarlo de nuevo (y/n): ').lower()
    while(answer != 'y' and answer != 'n'):
        answer = input(f'Su respuesta {answer} es ¡INVÁLIDA!, favor de intentarlo de nuevo (y/n): ')
```
