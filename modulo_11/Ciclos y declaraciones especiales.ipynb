# ciclos y declaraciones especiales

```python
import time
def countdown(time_sec):
    while time_sec:
        mins, secs = divmod(time_sec, 60)
        timeformat = '{:02d}:{:02d}'.format(mins, secs)
        print(timeformat, end='\r')
        time.sleep(1)
        time_sec -= 1
        print("stop")
# countdown(10)

def countup(time_sec):
    cont = 0
    cont_dos = 0
    min = 0
    while cont < time_sec+1:
        print(f'{min:02d}:{cont_dos:02d}')
        time.sleep(1)
        cont += 1
        cont_dos += 1
        if(cont_dos > 59):
            cont_dos = 0
            min += 1 
    print('Hemos llegado al fina lde la partida! ')
countup(70)
```
