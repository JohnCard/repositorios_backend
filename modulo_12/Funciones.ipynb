```python
"""
Actividad_modulo_12.ipynb

Automatically generated by Colaboratory.

Original file is located at
  https://colab.research.google.com/drive/1brLZoxViZBgRKB2leN6HzI8kOm3DhAMN
"""

import time

super_market = {
  '1.- Bolsa de papas': 10,
  '2.- Chocolate': 6,
  '3.- Refresco': 8,
  '4.- Pieza de pan': 4,
  '5.- Palomitas': 8,
  '6.- Paquete de galletas': 12,
  '7.- Paquete de papel': 18,
  '8.- Frasco de crema de cacahuate': 12
}

super_market_two = {
  '1.- Pastel': 20,
  '2.- PC': 100,
  '3.- Mochila': 30,
  '4.- Carne asada': 22,
  '5.- Celular': 60,
  '6.- TV': 70,
  '7.- Escoba': 14,
  '8.- Bocina': 25
}

super_market_three = {
  '1.- Lentes': 22,
  '2.- Envase de yogurt': 12,
  '3.- Pierna de jamon': 18,
  '4.- Juguete': 14,
  '5.- Jabón': 5,
  '6.- Botella de agua': 6,
  '7.- Cereal': 16,
  '8.- Frascos con granola': 8
}

def list_items_available(list_items,num):
  print(f'Lista de productos disponibles en el mercado número {num}')
  cont = 0
  for k,v in list_items.items():
    cont += 1
    print(f'{cont}.- Articulo: {k[4:]} // Precio del articulo {k[4:]}: ${v}')

def add_to_cart(list_items,item,num):
  cont = -1
  for it in list_items:
    cont += 1
    if num == str(it[0]):
      list_items[cont][1] += item[1]

def take_off(list_items,item,num):
  cont = -1
  for it in list_items:
    cont += 1
    if num == str(it[0]):
      list_items[cont][1] -= item[1]

def concatenar(list_items):
  cadena = ''
  for item in list_items:
    cadena += item[0]
  return cadena

def returning_message_market(chart):
  if chart == '1':
    message = '1.- Bolsa de papas'
  elif chart == '2':
    message = '2.- Chocolate'
  elif chart == '3':
    message = '3.- Refresco'
  elif chart == '4':
    message = '4.- Pieza de pan'
  elif chart == '5':
    message = '5.- Palomitas'
  elif chart == '6':
    message = '6.- Paquete de galletas'
  elif chart == '7':
    message = '7.- Paquete de papel'
  elif chart == '8':
    message = '8.- Frasco de crema de cacahuate'
  return message

def returning_message_second_market(chart):
  if chart == '1':
    message = '1.- Pastel'
  elif chart == '2':
    message = '2.- PC'
  elif chart == '3':
    message = '3.- Mochila'
  elif chart == '4':
    message = '4.- Carne asada'
  elif chart == '5':
    message = '5.- Celular'
  elif chart == '6':
    message = '6.- TV'
  elif chart == '7':
    message = '7.- Escoba'
  elif chart == '8':
    message = '8.- Bocina'
  return message

def returning_message_third_market(chart):
  if chart == '1':
    message = '1.- Lentes'
  elif chart == '2':
    message = '2.- Envase de yogurt'
  elif chart == '3':
    message = '3.- Pierna de jamon'
  elif chart == '4':
    message = '4.- Juguete'
  elif chart == '5':
    message = '5.- Jabón'
  elif chart == '6':
    message = '6.- Botella de agua'
  elif chart == '7':
    message = '7.- Cereal'
  elif chart == '8':
    message = '8.- Frascos con granola'
  return message

def get_my_account_info(list_items):
  
  referencia = input('Digite 1) para referenciarse del mercado 1, 2) para el mercado dos y 3) para el mercado 3: ')
  if referencia == '1':
    list_items_available(super_market,'1')
  elif referencia == '2':
    list_items_available(super_market_two,'2')
  elif referencia == '3':
    list_items_available(super_market_three,'3')
  
  def listening_my_orders(function):
    my_list = []
    def returning_sub_list(num):
      sub_list = []
      sub_list.append(function(num))
      sub_list.append(item[1])
      return sub_list
    for item in list_items:
      if str(item[0]) == '1':
        my_list.append(returning_sub_list('1'))
      elif str(item[0]) == '2':
        my_list.append(returning_sub_list('2'))
      elif str(item[0]) == '3':
        my_list.append(returning_sub_list('3'))
      elif str(item[0]) == '4':
        my_list.append(returning_sub_list('4'))
      elif str(item[0]) == '5':
        my_list.append(returning_sub_list('5'))
      elif str(item[0]) == '6':
        my_list.append(returning_sub_list('6'))
      elif str(item[0]) == '7':
        my_list.append(returning_sub_list('7'))
      elif str(item[0]) == '8':
        my_list.append(returning_sub_list('8'))
    return my_list
  
  def returning_state():
    if referencia == '1':
      my_list = listening_my_orders(returning_message_market)
    elif referencia == '2':
      my_list = listening_my_orders(returning_message_second_market)
    elif referencia == '3':
      my_list = listening_my_orders(returning_message_third_market)
    
    return my_list
  
  def sum_precios():
    cont = 0
    if referencia == '1':
      for item in list_items:
        message = returning_message_market(item[0])
        cont += item[1]*super_market[message]
    elif referencia == '2':
      for item in list_items:
        message = returning_message_second_market(item[0])
        cont += item[1]*super_market_two[message]
    elif referencia == '3':
      for item in list_items:
        message = returning_message_third_market(item[0])
        cont += item[1]*super_market_three[message]
    return f'En total usted en este preciso momento debe pagar ${cont}'

  def subtract_cart():
    print(f'Estado actual de su carrito de compras:')
    list_state = returning_state()
    for item in list_state:
      print(f'Producto: {item[0]}; Cantidad: {item[1]}')
    print(f'''Ahora usted debe digitar una lista de listas donde cada lista contendrá a la izquierda el índice del producto, y despues de 
la coma, insertar la cantidad de veces que desea retirarlo, aqui un ejemplo: si me referenciara del mercado "1":
[[1,3],[4,5]] que significaría retirar 3 bolsas de papas y 5 piezas de pan!''')
    quitar_carrito = eval(input('Digite la lista con los índices al lado de sus cantidades a retirar: '))
    for item in quitar_carrito:
      if '1' == str(item[0]):
        take_off(list_items,item,'1')
      elif '2' == str(item[0]):
        take_off(list_items,item,'2')
      elif '3' == str(item[0]):
        take_off(list_items,item,'3')
      elif '4' == str(item[0]):
        take_off(list_items,item,'4')
      elif '5' == str(item[0]):
        take_off(list_items,item,'5')
      elif '6' == str(item[0]):
        take_off(list_items,item,'6')
      elif '7' == str(item[0]):
        take_off(list_items,item,'7')
      elif '8' == str(item[0]):
        take_off(list_items,item,'8')
    new_list = returning_state()
    print('Estado nuevo de su carrito:')
    for item in new_list:
      print(f'{item[0]} // Cantidad: {item[1]}')
    return 'Gracias por su visita, hasta luego, que la fuerza lo acompañe!'

  def add_cart():
    print(f'Estado actual de su carrito de compras:')
    list_state = returning_state()
    for item in list_state:
      print(f'Producto: {item[0]}; Cantidad: {item[1]}')
    print(f'''Ahora usted debe digitar una lista de listas donde cada lista contendrá a la izquierda el índice del producto, y despues de 
la coma, insertar la cantidad de veces que desea repetirlo, aqui un ejemplo: si me referenciara del mercado "1":
[[1,3],[4,5]] que significaría 3 bolsas de papas y 5 piezas de pan!''')
    agregar_carrito = eval(input('Digite una lista con los índices y sus cantidades a agregar: '))
    for item in agregar_carrito:
      if '1' == str(item[0]):
        add_to_cart(list_items,item,'1')
        if '1' not in concatenar(list_items):
          list_items.append(item)
      elif '2' == str(item[0]):
        add_to_cart(list_items,item,'2')
        if '2' not in concatenar(list_items):
          list_items.append(item)
      elif '3' == str(item[0]):
        add_to_cart(list_items,item,'3')
        if '3' not in concatenar(list_items):
          list_items.append(item)
      elif '4' == str(item[0]):
        add_to_cart(list_items,item,'4')
        if '4' not in concatenar(list_items):
          list_items.append(item)
      elif '5' == str(item[0]):
        add_to_cart(list_items,item,'5')
        if '5' not in concatenar(list_items):
          list_items.append(item)
      elif '6' == str(item[0]):
        add_to_cart(list_items,item,'6')
        if '6' not in concatenar(list_items):
          list_items.append(item)
      elif '7' == str(item[0]):
        add_to_cart(list_items,item,'7')
        if '7' not in concatenar(list_items):
          list_items.append(item)
      elif '8' == str(item[0]):
        add_to_cart(list_items,item,'8')
        if '8' not in concatenar(list_items):
          list_items.append(item)
    new_list = returning_state()
    print('Estado nuevo de su carrito:')
    for item in new_list:
      print(f'{item[0]} // Cantidad: {item[1]}')
    return 'Gracias por su visita, hasta luego, que la fuerza lo acompañe!'

  print('''Digita a) para ver la sumatoria total de todas sus compras, b) para quitarle algo al carrito
ó c) para agregarle algo a tu carrito de compras:''')
  inciso = input('Digite su respuesta justo aqui (a/b/c): ').lower()
  if inciso == 'a':
    print(sum_precios())
  elif inciso == 'b':
    print(subtract_cart())
  elif inciso == 'c':
    print(add_cart())


# Ejemplo de uso de estas funciones
carrito_mercado_tres = [['2',3],['4',2],['6',4],['8',3]]
get_my_account_info(carrito_mercado_tres)
```
